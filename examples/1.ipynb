```python
import cv2
from google.colab.patches import cv2_imshow # for image display
from skimage import io

url = "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"

image = io.imread(url)      

image_2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
final_frame = cv2.hconcat((image, image_2))
#cv2_imshow(final_frame)
#cv2.imshow(image) does not work in colab
io.imshow(final_frame)
io.show()
print(image.shape)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
im2 = 255 - gray_image
cv2_imshow(im2)
```

```python
# Another transform of the image, after adding a constant, 
# all the pixels become brighter and a hazing-like effect of the image is generated
im3 = (100.0/255)*gray_image + 100
cv2_imshow(im3)
```

```python
# The lightness level of the gray_image decreases after this step
im4 = 255.0*(gray_image/255.0)**2
cv2_imshow(im4)
```

```python
flipped_code_0=cv2.flip(image,0) # vertical flip
cv2_imshow(flipped_code_0)
```

```python
flipped_code_1=cv2.flip(image,1) # horizontal flip
cv2_imshow(flipped_code_1)
```

```python
transposed=cv2.transpose(image)
cv2_imshow(transposed)
```

```python
for i in range(0,3):
   min_value, max_value, min_location, max_location=cv2.minMaxLoc(image[:,:,i])
   print("min {} is at {}, and max {} is at {}".format(min_value, min_location, max_value, max_location))
```
